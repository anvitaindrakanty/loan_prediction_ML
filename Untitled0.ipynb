```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler, OneHotEncoder
from sklearn.compose import ColumnTransformer
from sklearn.naive_bayes import GaussianNB
from sklearn.linear_model import LogisticRegression
from sklearn.ensemble import RandomForestClassifier, BaggingClassifier, GradientBoostingClassifier # New Imports
from sklearn.tree import DecisionTreeClassifier
from sklearn.neighbors import KNeighborsClassifier
from sklearn.metrics import accuracy_score, f1_score, confusion_matrix, classification_report
from sklearn.pipeline import Pipeline
import os

# --- Configuration ---
FILE_PATH = "/content/sample_data/loan_approval_dataset.csv"
RANDOM_STATE = 42
TRAINING_SIZE = 0.7
VALIDATION_SIZE = 0.1
TEST_SIZE = 0.2

# Check for XGBoost availability (requires separate installation)
try:
    from xgboost import XGBClassifier
    XGB_AVAILABLE = True
    print("XGBoost library found and will be included in modeling.")
except ImportError:
    XGB_AVAILABLE = False
    print("XGBoost library not found. Skipping XGBoost model. (Install with: pip install xgboost)")


# -----------------------------------------------------------
# 1. Data Preprocessing and Visualization (10 points)
# -----------------------------------------------------------

print("--- Step 1: Data Preprocessing and Visualization ---")

# Load the dataset
try:
    df = pd.read_csv(FILE_PATH)
except FileNotFoundError:
    print(f"Error: File not found at {FILE_PATH}. Please ensure the file is correctly uploaded.")
    exit()

# Clean column names
df.columns = df.columns.str.strip()

# Store original data for visualization later
df_original = df.copy()

# Drop 'loan_id'
df = df.drop('loan_id', axis=1)

# Encode the target variable: 'Approved' -> 1, 'Rejected' -> 0
df['loan_status'] = df['loan_status'].str.strip().map({'Approved': 1, 'Rejected': 0})
print("\nTarget variable encoded: Approved=1, Rejected=0")

# Define features (X) and target (y)
X = df.drop('loan_status', axis=1)
y = df['loan_status']

# Identify features for preprocessing
numerical_features = X.select_dtypes(include=['int64']).columns.tolist()
categorical_features = X.select_dtypes(include=['object']).columns.tolist()

# Define preprocessing steps
numerical_transformer = StandardScaler()
categorical_transformer = OneHotEncoder(handle_unknown='ignore')

preprocessor = ColumnTransformer(
    transformers=[
        ('num', numerical_transformer, numerical_features),
        ('cat', categorical_transformer, categorical_features)
    ],
    remainder='passthrough'
)

print("Data preprocessing (StandardScaler and OneHotEncoder) defined.")

# --- Visualization: Correlation Heatmap ---
print("\nGenerating Correlation Heatmap for numerical features...")

# Select all integer columns (including the 0/1 encoded target)
numerical_cols_for_corr = df.select_dtypes(include=['int64']).columns.tolist()

plt.figure(figsize=(12, 10))
correlation_matrix = df[numerical_cols_for_corr].corr()
sns.heatmap(correlation_matrix, annot=True, fmt=".2f", cmap='coolwarm', cbar=True)
plt.title('Correlation Matrix of Numerical Features and Loan Status')
plt.show()


# --- Other Visualizations ---

# 1. Distribution of the Target Variable (Loan Status)
plt.figure(figsize=(6, 4))
sns.countplot(x='loan_status', data=df)
plt.title('Loan Approval Status Distribution (0=Rejected, 1=Approved)')
plt.xticks([0, 1], ['Rejected', 'Approved'])
plt.ylabel('Count')
plt.xlabel('Loan Status')
plt.show()

# 2. Relationship between Education and Loan Status
plt.figure(figsize=(8, 5))
sns.countplot(x='education', hue='loan_status', data=df)
plt.title('Loan Status by Education Level')
plt.xlabel('Education')
plt.ylabel('Count')
plt.legend(title='Loan Status', labels=['Rejected', 'Approved'])
plt.show()

# 3. Distribution of CIBIL Score
plt.figure(figsize=(10, 5))
sns.histplot(df, x='cibil_score', hue='loan_status', bins=20, kde=True, palette='viridis')
plt.title('CIBIL Score Distribution by Loan Status')
plt.xlabel('CIBIL Score')
plt.ylabel('Density/Count')
plt.show()


# -----------------------------------------------------------
# 2. Split the Data Set (70% Train, 10% Validation, 20% Test)
# -----------------------------------------------------------

print("\n--- Step 2: Data Splitting (70/10/20) ---")

# First split: Separate out 20% for the final test set
X_temp, X_test, y_temp, y_test = train_test_split(
    X, y, test_size=TEST_SIZE, random_state=RANDOM_STATE, stratify=y
)

# Calculate the validation size proportion relative to the temporary set
val_size_relative = VALIDATION_SIZE / (TRAINING_SIZE + VALIDATION_SIZE)

# Second split: Divide the remaining data into 70% training and 10% validation
X_train, X_val, y_train, y_val = train_test_split(
    X_temp, y_temp, test_size=val_size_relative, random_state=RANDOM_STATE, stratify=y_temp
)

print(f"X_train shape: {X_train.shape}, y_train shape: {y_train.shape} ({TRAINING_SIZE*100:.0f}%)")
print(f"X_val shape: {X_val.shape}, y_val shape: {y_val.shape} ({VALIDATION_SIZE*100:.0f}%)")
print(f"X_test shape: {X_test.shape}, y_test shape: {y_test.shape} ({TEST_SIZE*100:.0f}%)")


# -----------------------------------------------------------
# 3. Classification Methods (50 points)
# -----------------------------------------------------------

print("\n--- Step 3: Classification Models and Training ---")

# Define the models (classifiers)
models = {
    'Logistic Regression': LogisticRegression(random_state=RANDOM_STATE, max_iter=1000),
    'Gaussian Naive Bayes': GaussianNB(),
    'Decision Tree': DecisionTreeClassifier(random_state=RANDOM_STATE),
    'K-Nearest Neighbors': KNeighborsClassifier(n_neighbors=5),

    # Bagging Methods
    'Random Forest': RandomForestClassifier(random_state=RANDOM_STATE, n_estimators=100),
    'Bagging (Decision Tree)': BaggingClassifier(
        estimator=DecisionTreeClassifier(random_state=RANDOM_STATE),
        n_estimators=50,
        random_state=RANDOM_STATE
    ),

    # Boosting Methods
    'Gradient Boosting': GradientBoostingClassifier(random_state=RANDOM_STATE, n_estimators=100),
}

# Add XGBoost conditionally (another Boosting method)
if XGB_AVAILABLE:
    models['XGBoost'] = XGBClassifier(random_state=RANDOM_STATE, use_label_encoder=False, eval_metric='logloss')

results = {}

# Train and evaluate each model using a Pipeline
for name, model in models.items():
    print(f"\nTraining {name}...")

    # Create a pipeline: Preprocessor -> Model
    pipeline = Pipeline(steps=[
        ('preprocessor', preprocessor),
        ('classifier', model)
    ])

    # Train the model on the training data
    pipeline.fit(X_train, y_train)

    # Evaluate the final performance on the unseen test set
    y_test_pred = pipeline.predict(X_test)
    test_accuracy = accuracy_score(y_test, y_test_pred)
    test_f1 = f1_score(y_test, y_test_pred)
    cm = confusion_matrix(y_test, y_test_pred)

    # Store results
    results[name] = {
        'Test_Accuracy': test_accuracy,
        'Test_F1_Score': test_f1,
        'Confusion_Matrix': cm,
        'Classification_Report': classification_report(y_test, y_test_pred, output_dict=True)
    }

    print(f"  Test Accuracy: {test_accuracy:.4f}")

# -----------------------------------------------------------
# 4. Results and Evaluation (30 points)
# -----------------------------------------------------------

print("\n--- Step 4: Final Results and Evaluation ---")

# Tabulate the main results
evaluation_summary = pd.DataFrame({
    'Model': results.keys(),
    'Accuracy': [res['Test_Accuracy'] for res in results.values()],
    'F1 Score': [res['Test_F1_Score'] for res in results.values()]
})

print("\n--- Evaluation Summary Table ---")
print(evaluation_summary.sort_values(by='F1 Score', ascending=False).to_markdown(index=False, floatfmt=".4f"))

# Detailed results and Confusion Matrix for the best model (based on F1-Score)
best_model_name = evaluation_summary.sort_values(by='F1 Score', ascending=False).iloc[0]['Model']
best_results = results[best_model_name]

print(f"\n--- Detailed Results for Best Model: {best_model_name} ---")

# Confusion Matrix Visualization
plt.figure(figsize=(6, 6))
sns.heatmap(best_results['Confusion_Matrix'], annot=True, fmt='d', cmap='Blues',
            xticklabels=['Rejected (0)', 'Approved (1)'], yticklabels=['Rejected (0)', 'Approved (1)'])
plt.title(f'Confusion Matrix for {best_model_name}')
plt.xlabel('Predicted Label')
plt.ylabel('True Label')
plt.show()

# Classification Report
print("\nClassification Report (Test Set):")
report_df = pd.DataFrame(best_results['Classification_Report']).transpose()
print(report_df.to_markdown(floatfmt=".4f"))

# -----------------------------------------------------------
# 5. Conclusion (10 points)
# -----------------------------------------------------------

print("\n--- Step 5: Conclusion ---")

# Determine which models were included for the conclusion
models_list_for_conclusion = list(models.keys())
models_str = ", ".join(models_list_for_conclusion)

print(f"""
Based on the analysis and model evaluation on the 20% test set, incorporating {len(models_list_for_conclusion)} different classification methods ({models_str}):

1. Data Preprocessing: The data was successfully preprocessed, and a ColumnTransformer was used for feature scaling (StandardScaler) and categorical encoding (OneHotEncoder).

2. Model Performance:
   - The comparison now includes three Bagging methods (Random Forest, Bagging Classifier) and two Boosting methods (Gradient Boosting, XGBoost if available). The **{best_model_name}** model achieved the highest performance metrics, with a Test Accuracy of **{best_results['Test_Accuracy']:.4f}** and an F1-Score of **{best_results['Test_F1_Score']:.4f}**.
   - The inclusion of explicit Bagging and Gradient Boosting provides a more comprehensive comparison of ensemble techniques.

3. Key Insights: The **Correlation Heatmap** confirms that **CIBIL Score** is the single most correlated feature with `loan_status`, showing a strong positive relationship (higher CIBIL = higher approval chance), followed by the asset values.

In summary, the classification problem was solved using a structured 70/10/20 train/validation/test split, and the best model chosen provided the optimal balance of precision and recall on the unseen test data.
""")
```
